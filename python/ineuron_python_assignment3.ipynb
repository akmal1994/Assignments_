### 1.1 Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce()

```python
def myreduce(number):
    
    l = []
    for i in range(1, number+1):
        l.append(i)
        
    y = sum(l)
    
    return l, y

print("Input")
number = int(input("Please enter the number:\n"))

result = myreduce(number)

print("Result:\n")
print(f"Natural numbers are: {result[0]}.")
print(f"Sum of the elements are: {result[1]}")
```

### 1.2 Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter()

```python
def odd_num(number):
    if number % 2 == 0:
        return False
    else:
        return True
    
def even_num(number):
    if number % 2 == 0:
        return True
    else:
        return False
    
def myfilter(function, list_num):
    
    l = []
    for i in list_num:
        if function(i) == True:
            l.append(i)
    return l
```

```python
lst = [1,2,3,4,5,6,7,8]
myfilter(odd_num, lst)
```

### 2. Implement List comprehensions to produce the following lists. Write List comprehensions to produce the following Lists
['A', 'C', 'A', 'D', 'G', 'I', ’L’, ‘ D’]
['x', 'xx', 'xxx', 'xxxx', 'y', 'yy', 'yyy', 'yyyy', 'z', 'zz', 'zzz', 'zzzz']
['x', 'y', 'z', 'xx', 'yy', 'zz', 'xxx', 'yyy', 'zzz', 'xxxx', 'yyyy', 'zzzz'] [[2], [3], [4], [3], [4], [5], [4], [5], [6]] [[2, 3, 4, 5], [3, 4, 5, 6], [4, 5, 6, 7], [5, 6, 7, 8]]
[(1, 1), (2, 1), (3, 1), (1, 2), (2, 2), (3, 2), (1, 3), (2, 3), (3, 3)]

```python
# ['x', 'xx', 'xxx', 'xxxx', 'y', 'yy', 'yyy', 'yyyy', 'z', 'zz', 'zzz', 'zzzz']

base_list = ['x', 'y', 'z']
new_list = [ele*num for ele in base_list for num in range(1, 5)]
print(base_list, "==>", new_list)
```

```python
base_list1 = ['x', 'y', 'z']
nw_list = []

for el in base_list1:
    for nm in range(1, 5):
        nw_list.append(el*nm)
        
print(base_list1, "==>", nw_list)
```

```python
# ['x', 'y', 'z', 'xx', 'yy', 'zz', 'xxx', 'yyy', 'zzz', 'xxxx', 'yyyy', 'zzzz']

new_list = [ele*num for num in range(1, 5) for ele in base_list]
print(base_list, "==>", new_list)
```

```python
nw_list = []
for nm in range(1,5):
    for el in base_list1:
        nw_list.append(nm*el)
        
print(base_list1, '==>', nw_list)
        
        
```

```python
# [[2], [3], [4], [3], [4], [5], [4], [5], [6]]

num_list = [2, 3, 4]
new_list = [[ele+num] for ele in num_list for num in range(3)]
print(num_list, "==>", new_list)
```

```python
no_list = [2,3,4]
nw_no_list = []
for elm in no_list:
    for no in range(3):
        nw_no_list.append([elm+no])
        
print(no_list, '==>', nw_no_list)
```

```python
# [[2, 3, 4, 5], [3, 4, 5, 6],[4, 5, 6, 7], [5, 6, 7, 8]]

num_list = [2, 3, 4, 5]
new_list = [[ele+num for ele in num_list] for num in range(4)]
print(num_list, "==>", new_list)
```
